```python
#The function make line graph
def lgraf(x,y):
    x.value_counts()
    dfneworder=pd.DataFrame()
    dfneworder["f"]=x.dt.date
    nw=pd.DataFrame()
    nw=dfneworder.value_counts().sort_index(ascending=True).to_frame(name="counts")
    plt.figure(figsize=(20,15))
    a=sns.lineplot(data=nw, x="f", y="counts")
    a.set_title(y, fontsize = 20)
    plt.show()
```

```python
lgraf(dfOrder["ordoro_import_date"],"The date order was imported to ordoro")
lgraf(dfOrder["tool_import_date"],"The date the order was imported in the Tool")
lgraf(dfOrder["max_shipping_date"],"Max shipping date is the max target shipping date for an order")
lgraf(dfOrder["max_delivery_date"],"Max shipping date is the max target delivery date for an order")
```
